# Train — SmallUNet σ=5

本番採用の標準訓練ノートブック。
- モデル: SmallUNet
- sigma: 5
- データ: `train/dataset/`
- 出力: `train/runs/`

CLIで実行する場合:
```bash
uv run python train/train.py --data-dir train/dataset --backbone smallunet --sigma 5 --epochs 50
```

````python
import sys
from pathlib import Path

# workspace root から実行することを想定
ROOT = Path(".").resolve()
TRAIN_DIR = ROOT / "train"
if str(TRAIN_DIR) not in sys.path:
    sys.path.insert(0, str(TRAIN_DIR))
````

````python
# ---- 設定 ----
DATA_DIR   = "train/dataset"
SAVE_DIR   = "train/runs"
BACKBONE   = "smallunet"
SIGMA      = 5.0
EPOCHS     = 50
BATCH_SIZE = 4
LR         = 1e-3
RESIZE     = (512, 512)
AUGMENT    = False
LOSS       = "mse"
CHECKPOINT = None  # 再開時はパスを指定
````

````python
import train as trainer
import argparse

args = argparse.Namespace(
    data_dir=DATA_DIR,
    save_dir=SAVE_DIR,
    backbone=BACKBONE,
    sigma=SIGMA,
    epochs=EPOCHS,
    batch_size=BATCH_SIZE,
    lr=LR,
    resize=list(RESIZE),
    augment=AUGMENT,
    loss=LOSS,
    checkpoint=CHECKPOINT,
)

trainer.main(args)
````
